```python
import time
import random

def perform_dance(name, moves, delay):
    print(f"\n--- 💃 Starting: {name} ---")
    for i in range(3, 0, -1):
        print(f"Ready in {i}...")
        time.sleep(0.6)
    
    for move in moves:
        print(f">> {move} <<")
        time.sleep(delay)
    print(f"--- ✨ End of {name} ✨ ---\n")

def main():
    # Define different dance routines
    dances = {
        "1": ("The Nay Nay", ["Step Forward", "Side Step", "Close Feet", "Step Back"], 1.5),
        "2": ("Harlem Shake", ["Point Up Right", "Point Down Left", "Hip Sway", "Spin!"], 0.8),
        "3": ("Renegade", ["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 1.2),
        "4": ("Supalonely", ["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 1.2),
        "5": ("WAP", ["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 1.2),
        "6": ("Say So", ["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 1.2),
        "7": ("Blinding Lights", ["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 1.2),
        "8": ("Tap In", ["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 1.2),
        "9": ("Savage", ["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 1.2),
    }
    while True:
        print("--- 🕺 MAIN MENU 💃 ---")
        print("1. The Nay Nay")
        print("2. Harlem Shake")
        print("3. Renegade")
        print("4. Supalonely")
        print("5. WAP")
        print("6. Say So")
        print("7. Blinding Lights")
        print("8. Tap In")
        print("9. Savage")
        print("10. Surprise Me! (Random)")
        print("11. Exit")
        
        choice = input("\nPick a dance number (1-11): ")

        if choice == "11":
            print("Thanks for dancing! See ya!")
            break
        elif choice == "10":
            # Pick a random dance from the dictionary
            name, moves, speed = random.choice(list(dances.values()))
            perform_dance(f"Surprise {name}", moves, speed)
        elif choice in dances:
            name, moves, speed = dances[choice]
            perform_dance(name, moves, speed)
        else:
            print("❌ Invalid choice! Please pick 1-11.")

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
import time
import threading # Used to keep the window from freezing during the dance

def perform_dance(name, moves, delay, label):
    label.config(text=f"💃 Starting: {name}...")
    root.update()
    time.sleep(1) # Quick prep
    
    for move in moves:
        label.config(text=f">> {move} <<")
        root.update()
        time.sleep(delay)
        
    label.config(text=f"✨ Finished {name}! ✨")

def start_thread(name, moves, delay, label):
    # Running the dance in a separate thread prevents the "Not Responding" error
    threading.Thread(target=perform_dance, args=(name, moves, delay, label), daemon=True).start()

# --- GUI SETUP ---
root = tk.Tk()
root.title("Dance Studio")
root.geometry("400x600")

# Dance Data
dances = {
    "Lottery (Renegade) - K Camp": (["Step Forward", "Side Step", "Close Feet", "Step Back"], 1.0),
    "The Git Up - Blanco Brown": (["Point Up Right", "Point Down Left", "Hip Sway", "Spin!"], 0.6),
    "Juju on that Beat - Zay Hilfigerrr": (["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 0.8),
    "Adderall - Popp Hunna": (["Step Forward", "Side Step", "Close Feet", "Step Back"], 1.0),
    "E85 - Don Toliver": (["Point Up Right", "Point Down Left", "Hip Sway", "Spin!"], 0.6),
    "Savage Love - Jason Derulo": (["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 0.8),
    "Savage - Megan Thee Stallion": (["Step Forward", "Side Step", "Close Feet", "Step Back"], 1.0),
    "Just Wanna Rock - Lil Uzi Vert": (["Point Up Right", "Point Down Left", "Hip Sway", "Spin!"], 0.6),
    "Say So - Doja Cat": (["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 0.8),
    "Toosie Slide - Drake": (["Step Forward", "Side Step", "Close Feet", "Step Back"], 1.0),
}

# UI Elements
tk.Label(root, text="🕺 SELECT YOUR DANCE 💃", font=("Arial", 14, "bold")).pack(pady=20)

# Status Display (This replaces the print statements)
status_label = tk.Label(root, text="Ready to dance...", font=("Courier", 12), fg="purple")
status_label.pack(pady=20)

# Create Buttons for the Top 3
for name, (moves, delay) in dances.items():
    btn = tk.Button(root, text=name, width=20, font=("Arial", 10),
                   command=lambda n=name, m=moves, d=delay: start_thread(n, m, d, status_label))
    btn.pack(pady=5)

root.mainloop()
```

```python
import tkinter as tk
from PIL import Image, ImageTk # New: for image handling
import time
import threading

def perform_dance(name, moves, delay, status_label, image_label):
    status_label.config(text=f"💃 Starting: {name}...")
    
    for move in moves:
        status_label.config(text=f">> {move} <<")
        
        # Try to load an image named after the move (e.g., "Step Forward.png")
        try:
            img = Image.open(f"images/{move}.png").resize((250, 250))
            photo = ImageTk.PhotoImage(img)
            image_label.config(image=photo, text="")
            image_label.image = photo # Important: Keep a reference!
        except:
            image_label.config(image='', text="[No Image Found]")
            
        time.sleep(delay)
        
    status_label.config(text=f"✨ Finished {name}! ✨")

def start_thread(name, moves, delay, s_label, i_label):
    threading.Thread(target=perform_dance, args=(name, moves, delay, s_label, i_label), daemon=True).start()

# --- GUI SETUP ---
root = tk.Tk()
root.title("Dance Studio")
root.geometry("500x800")

tk.Label(root, text="🕺 DANCE STUDIO 💃", font=("Arial", 14, "bold")).pack(pady=10)

# New: Label to display the dance images
image_display = tk.Label(root, text="Images appear here", bg="lightgrey", width=30, height=12)
image_display.pack(pady=10)

status_label = tk.Label(root, text="Ready to dance...", font=("Courier", 12), fg="purple")
status_label.pack(pady=10)

# Dance Data
dances = {
    "Lottery (Renegade) - K Camp": (["Step Forward", "Side Step", "Close Feet", "Step Back"], 1.0),
    "The Git Up - Blanco Brown": (["Point Up Right", "Point Down Left", "Hip Sway", "Spin!"], 0.6),
    "Juju on that Beat - Zay Hilfigerrr": (["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 0.8),
    "Adderall - Popp Hunna": (["Step Forward", "Side Step", "Close Feet", "Step Back"], 1.0),
    "E85 - Don Toliver": (["Point Up Right", "Point Down Left", "Hip Sway", "Spin!"], 0.6),
    "Savage Love - Jason Derulo": (["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 0.8),
    "Savage - Megan Thee Stallion": (["Step Forward", "Side Step", "Close Feet", "Step Back"], 1.0),
    "Just Wanna Rock - Lil Uzi Vert": (["Point Up Right", "Point Down Left", "Hip Sway", "Spin!"], 0.6),
    "Say So - Doja Cat": (["Lock Elbows", "Pivot Head", "Mechanical Slide", "Beep-Boop Pose"], 0.8),
    "Toosie Slide - Drake": (["Step Forward", "Side Step", "Close Feet", "Step Back"], 1.0),
}

for name, (moves, delay) in dances.items():
    # We now pass both the status label and the image label
    btn = tk.Button(root, text=name, width=30, 
                   command=lambda n=name, m=moves, d=delay: start_thread(n, m, d, status_label, image_display))
    btn.pack(pady=5)

root.mainloop()
```
